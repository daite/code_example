```python
import pandas as pd; import numpy as np
from pandas import DataFrame, Series
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
fig, axes = plt.subplots(2,2)
ax1 = axes[0,0]
ax2 = axes[0,1]
ax3 = axes[1,0]
ax4 = axes[1,1]
ax1.plot(np.random.randn(100).cumsum(), color='purple', label='ax1', marker='o')
ax1.legend(loc='best')
ax2.plot(np.random.randn(100).cumsum(), color='blue', label='ax2', marker='.')
ax2.legend(loc='best')
ax3.plot(np.random.randn(100).cumsum(), color='red', label='ax3', marker='>')
ax3.legend(loc='best')
ax4.plot(np.random.randn(100).cumsum(), color='green', label='ax4', alpha=0.7,
        marker='<')
ax4.legend(loc='best')
```
